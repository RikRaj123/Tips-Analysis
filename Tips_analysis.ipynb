```python
import numpy
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data1 = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
data1
```

```python
data1.info()
```

```python
sns.jointplot(x='time',y='tip',data=data1)
```

```python
sns.jointplot(x='day',y='tip',data=data1)
```

```python
data2=data1.loc[data1["day"]=='Sat']
sns.jointplot(x='time',y='tip',data=data2)
```

```python
sns.jointplot(x='sex',y='tip',data=data2)
```

**Highest tip are provided at Saturday dinner time.**

```python
print("A graph analysing males and females who tips with respect to total bill")
sns.scatterplot(x='total_bill',y='tip',data=data1,hue='sex')
```

***Males pay more tips than females.***

```python
print("Analysing the time with maximum occupency")
sns.countplot(x='time',data=data1)
```

**Dinners have more occupency in a restaurant than Lunch.**

<p><b>According to the analysis:-

*   Highest Tips are provided on Saturday
*   Males pay more tips than Females
*   Dinner Time has more occupency in a restaurant than lunch time.</b></p>
